```python
import pandas as pd

# The correct function name is pd.Series
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=["d", "b", "a", "c"])
print(obj)
obj2 = obj.reindex(["a", "b", "c", "d", "e"])
print(obj2)
```

```python
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
print(obj3)
obj4 = obj3.reindex(range(6), method="ffill")
print(obj4)
```

```python
obj3 = pd.Series(["blue", "purple", "yellow"], index=[0, 2, 4])
print(obj3)
obj4 = obj3.reindex(range(6), method="ffill")
print(obj4)

obj5 = obj3.reindex (range(6), method="bfill")
print("\nMethod'bfill':")
print(obj5)

obj6 = obj3.reindex(range(6), fill_value='red')
print("\nFill value:")
print(obj6)

obj7 = pd.Series ([1,2,3], index =[0,2,5])
obj8 = obj7.reindex(range(7), method="ffill", limit=1)
print("\nLimit:")
print(obj8)
```

```python
import pandas as pd
import numpy as np

obj = pd.Series([7, -5, 7, 3, 2, 0, 4])


df_rank = obj.rank()
print("Rank\n", df_rank)


df_rank_method_first = obj.rank(method='first')
print("\nrank-Method=first\n", df_rank_method_first)


df_rank_ascending = obj.rank(ascending=False, method='max')
print("\nRank-Ascending\n", df_rank_ascending)


frame = pd.DataFrame({'b': [4.3, 7, -3, 2], 'a': [0, 1, 0, 1],
                      'c': [-2, 5, 8, -2.5]})
print("\nDataFrame\n", frame)


df_rank_dataframe = frame.rank(axis='columns')
print("\nRank over columns\n", df_rank_dataframe)
```
